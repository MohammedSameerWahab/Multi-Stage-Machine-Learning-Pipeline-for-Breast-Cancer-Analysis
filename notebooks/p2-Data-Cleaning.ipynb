```python
# Create a new DataFrame without the missing rows
df_cleaned = df.dropna()

print("DataFrame after dropping rows with any missing value:")
df_cleaned
```

```python
# 1. Define the list of string values you want to remove.
values_to_remove = ["Not Performed", "Performed but Not Available", "Indeterminate", "Not Available"]

# 2. Create a boolean mask.
# This checks every cell. If a cell contains one of the values_to_remove, it becomes True.
# .any(axis=1) then marks the entire row as True if it contains at least one True value.
rows_to_drop_mask = df_cleaned.isin(values_to_remove).any(axis=1)

# 3. Filter the DataFrame.
# The '~' (tilde) operator inverts the boolean mask, so you keep only the rows that are False (i.e., rows without the specified values).
df_further_cleaned = df_cleaned[~rows_to_drop_mask]


# --- Display the Result ---
print("Cleaned DataFrame:")
print(df_further_cleaned)
```

```python
#identifying garbage values
for i in df_further_cleaned.select_dtypes(include="object").columns:
    print(df_further_cleaned[i].value_counts())
    print("***"*10)
```

```python
# This creates a new DataFrame that excludes rows where 'HER2.Final.Status' is 'Equivocal'.
df_final = df_further_cleaned[df_further_cleaned['HER2.Final.Status'] != 'Equivocal']


# --- Display the Result ---
print("DataFrame After Removal:")
print(df_final)
```

```python
df_final.head
```

```python
df_final = df_final.drop('vital.status', axis=1)

# --- Display the Result ---
print("DataFrame After Removing Column:")
print(df_final)
```

```python
# 1. Define the list of conditions in order of priority.
# Each condition is a boolean mask that returns True/False for each row.
conditions = [
    # Condition 1: HER2-enriched
    (df_final['HER2.Final.Status'] == 'Positive'),

    # Condition 2: Triple-Negative
    (df_final['ER.Status'] == 'Negative') &
    (df_final['PR.Status'] == 'Negative') &
    (df_final['HER2.Final.Status'] == 'Negative'),
]

# 2. Define the list of corresponding values for each condition.
choices = [
    'HER2-enriched',
    'Triple-Negative'
]

# 3. Use numpy.select() to create the new column.
# It checks conditions in order. If a condition is met, it assigns the corresponding choice.
# If no conditions are met, it assigns the default value.
df_final['subtype'] = np.select(conditions, choices, default='HR-positive/HER2-negative')


# --- Display the Final Result ---
print("DataFrame After Creating Subtype:")
print(df_final)
```
